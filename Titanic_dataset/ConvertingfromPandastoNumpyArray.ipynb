__Converting from a Pandas Series to a Numpy Array__

```python
'''The result is a 1-dimensional array. You can tell since there's only one set of brackets and it only expands across the page (not down as well).'''
import pandas as pd
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
print(df['Fare'].values)

#The values attribute of a Pandas Series give the data as a numpy array.
```

```python
'''This is a 2-dimensional numpy array.You can tell because thereâ€™s two sets of brackets, and it expands both across the page and down.'''

import pandas as pd
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
print(df[['Pclass', 'Fare', 'Age']].values)
#The values attribute of a Pandas DataFrame give the data as a 2d numpy array.
```

__Numpy Shape Attribute__

```python
import pandas as pd
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
arr = df[['Pclass', 'Fare', 'Age']].values
print(arr.shape)

# This result means we have 887 rows and 3 columns.
# Use the shape attribute to find the number of rows and number columns for a Numpy array. You can also use the shape attribute on a pandas DataFrame (df.shape).
```

__Select from a Numpy Array__

```python
import pandas as pd
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
arr = df[['Pclass', 'Fare', 'Age']].values

print(arr[0, 1])    # We can select a single element from a numpy array with the following.This will be the 2nd column of the 1st row. Hence (row, column). Remember that we start countinf at 0.
                    # SO it';; be the Fare (column=1) of the 1st passenger (row=0), or 7.5 
print(arr[0])       # We can also select a single row, for example, the whole row of the first passenger. anwer: [Pclass, Fare, Age] or [3. 7.25 22.]

print(arr[:,2])     #To select a single column (in this case the Age column), we have to use some special syntax : 
                    #The syntax : can be interpreted as we're taking all of the rowsm but just to column at index 2.
```

__Masking__

```python
import pandas as pd

df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')

# take first 10 values for simplicity
arr = df[['Pclass', 'Fare', 'Age']].values[:10]

mask = arr[:, 2] < 18      # We create what we call a mask first. This is an array of boolean values. (True/False) of whether the passenger is a child or not.

#print(mask)
print(arr[mask])  

# or can write like this below.
print(arr[arr[:, 2] < 18])
```

__Summing and Counting__

```python
import pandas as pd

df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
arr = df[['Pclass', 'Fare', 'Age']].values
mask = arr[:, 2] < 18

print(mask.sum())  #using attribute .sum(), we can count how many children or passengers below 18 years old

print((arr[:, 2] < 18).sum())
```
